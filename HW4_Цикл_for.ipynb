**1. Цикл for - базовий**

Використовуючи цикл `for`, виведіть числа від 1 до 7 включно використовуючи функцію range.


```python
for i in range(1, 8):
  print(i)
```


---

**2. Цикл for по списку**

Дано список:

```python
names = ["Олег", "Анна", "Іван", "Марія"]
```


Використовуючи цикл, виведіть кожне ім'я з привітанням: "Привіт, [ім'я]!"


```python
for i in range(len(names)):
  print(f"Привіт, {names[i]}!")
```

**3. Цикл з enumerate**

Дано список оцінок:

```python
grades = [85, 92, 78, 95, 88]
```


Використовуючи `enumerate`, виведіть кожну оцінку з номером: "Студент 1: 85 балів"


```python
for i in enumerate(grades):
  print(f"Студент {i[0] + 1}: {i[1]} балів")
```

**4. Сума парних чисел**

Напишіть програму, яка обчислює суму всіх парних чисел від 1 до 20.

```python
result_sum = 0

for i in range(1, 21):
  if i % 2 == 0:
    result_sum += i

print(result_sum )
```

```python
### another way:

result_sum = sum([i for i in range(1, 21) if i%2 == 0])
print(result_sum)
```
